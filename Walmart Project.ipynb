```python
df["unit_price"] = df["unit_price"].str.replace("$","")
df["unit_price"]= df["unit_price"].astype(float)
```

```python
df["total"]= df["unit_price"]* df["quantity"]
df["total"]
```

```python
df["profit"]= df["total"]* df["profit_margin"]
```

```python
df["cost"] = df["total"] - df["profit"] 
```

```python
df["date"] = pd.to_datetime(df["date"], format = "%d/%m/%y") 
```

```python
df["time"] = pd.to_datetime(df["time"], format= "%H:%M:%S")
df["hour"] = df["time"].dt.hour
```

```python
df["time"] = df["time"].dt.time
```

```python
category_sales= df.groupby("category")["total"].sum().sort_values(ascending= False)
category_sales
```

```python
bins= [3,5,8,10]
labels= ["low","Medium","High"]

df["rating_group"] = pd.cut(
    df["rating"],
    bins=bins,
    labels=labels,
    right=True,
    include_lowest=True
)

df[df["rating"] == 3.0]
df[df["rating"] == 10.0]
```

```python
df["year"]= df["date"].dt.year
df["month"]= df["date"].dt.month
df["month_name"]= df["date"].dt.month_name()
df["week"] = df["date"].dt.day_name()
df["week_number"] = df["date"].dt.isocalendar().week
```

```python
df[["cost","profit","total"]] = df[["cost", "profit","total"]].round(2)
```

```python
from sqlalchemy import create_engine
import urllib.parse

# 1. Database connection details
username = "root"
password = "yourpassword"
host = "localhost"
port = "3306"
database = "walmart_project"

# 2. Safely encode the password (essential due to '@')
safe_password = urllib.parse.quote_plus(password)

# 3. Create the engine
# Note: Ensure you have pymysql installed (pip install pymysql)
engine = create_engine(f"mysql+pymysql://{username}:{safe_password}@{host}:{port}/{database}")

# 4. Write DataFrame to MySQL
# 'replace' will overwrite the table; use 'append' to add to it
df.to_sql("walmart_sales", con=engine, if_exists="replace", index=False)

# 5. Read back a sample
sample_df = pd.read_sql("SELECT * FROM walmart_sales LIMIT 5;", con=engine)
print(sample_df)
```
